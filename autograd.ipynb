```python
import numpy as np
import torch
```

```python
x = torch.randn(3, requires_grad=True)
x
```

```python
y = x+2
y
```

```python
z = y*y*2
z
```

```python
z = z.mean()
z
```

```python
z.backward()
```

```python
x.grad
```

```python
x.requires_grad_(False)
x
```

```python
y = x.detach()
y
```

```python
with torch.no_grad():
    y = x+2
    print(y)
```

```python
weights = torch.ones(4, requires_grad=True)

for epoch in range(3):
  model_output = (weights*3).sum()

  model_output.backward()

  print(weights.grad)
```

```python
weights.grad.zero_()
```

```python
optimizer = torch.optim.SGD(weights, lr=0.01)
optimizer.step()
optimizer.zero_grad()
```
